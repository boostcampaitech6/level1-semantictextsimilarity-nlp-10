## 데이터 전처리

```python
lis1 = list(b['sentence_1'])
lis2 = list(b['sentence_2'])

len(lis1)
```

### re 라이브러리로 특수기호, ㅋㅋㅋ, ㅎㅎㅎ 등 자음 제거하기

```python
# 특수기호, ㅎㅋ같은 자음 제거 함수
import re
def extract_word(text):
    hangul = re.compile('[^가-힣a-zA-Z]')
    result = hangul.sub(' ', text)
    return result.lower()
```

```python
lis_re1 = []
lis_re2 = []

for q,w in zip(lis1, lis2):
    lis_re1.append(re.sub(' +', ' ', extract_word(q)))
    lis_re2.append(re.sub(' +', ' ', extract_word(w)))
lis_re1[:15] # 특수기호, ㅎㅎㅎ 제거됨.
```

### stopword 불러오기

```python
f = open("stopwords_yhe.txt", 'r', encoding = 'utf-8') #경로는 수정하세용
lines = f.readlines()
stopwords = []
for line in lines:
    line = line.replace('\n', '')
    stopwords.append(line)
f.close()
```

```python
def filter_stopwords(tokenized_text : list, stopwords_2 : list):
  tokenized_filtered = []

  for i in tokenized_text:
    if i not in stopwords and i not in stopwords_2:
        tokenized_filtered.append(i)

  return tokenized_filtered
```

```python
lis_st_re1 = []
lis_st_re2 = []

for q,w in zip(lis_re1, lis_re2):
    lis_st_re1.append(' '.join(filter_stopwords(q.strip().split(), stopwords)))
    lis_st_re2.append(' '.join(filter_stopwords(w.strip().split(), stopwords)))
lis_st_re1[:15]# 스탑워드 제거
```
