## output 사후 분석

```python
koelectra_output = pd.read_csv('output3_koelectra.csv')
koelectra_output.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
test = pd.merge(test, koelectra_output, on = 'id', how = 'left')
```

```python
test.head()
```

```python
plt.hist(test.target)  # 결과 분포는 고르게 나옴! 
```

```python
# target 값 분포
import matplotlib.pyplot as plt

data_list = [['nsmc-sampled','nsmc-rtt'],['slack-sampled','slack-rtt'],['petition-sampled','petition-rtt']]
for i,source in enumerate(data_list): 
    plt.figure(figsize=(12, 4))  # Figure 크기 설정

    plt.subplot(1, 2, 1)
    plt.hist(test[(test.source == data_list[i][0])].target, bins=20, alpha=0.5, color='blue')
    plt.xlabel('Label')
    plt.ylabel('Frequency')
    plt.title(f'test {data_list[i][0]} Histogram')

    plt.subplot(1, 2, 2)
    plt.hist(test[(test.source == data_list[i][1])].target, bins=20, alpha=0.5, color='red')
    plt.xlabel('Label')
    plt.ylabel('Frequency')
    plt.title(f'test {data_list[i][1]} Histogram')


    plt.tight_layout()  
    plt.show()
```

```python
# train과 비교

data_list = [['nsmc-sampled','nsmc-rtt'],['slack-sampled','slack-rtt'],['petition-sampled','petition-rtt']]

for i,source in enumerate(data_list): 
    plt.figure(figsize=(12, 4))  # Figure 크기 설정

    plt.subplot(1, 2, 1)
    plt.hist(train[(train.source == data_list[i][0])].label, bins=20, alpha=0.5, color='blue')
    plt.xlabel('Label')
    plt.ylabel('Frequency')
    plt.title(f'train {data_list[i][0]} Histogram')

    plt.subplot(1, 2, 2)
    plt.hist(train[(train.source == data_list[i][1])].label, bins=20, alpha=0.5, color='red')
    plt.xlabel('Label')
    plt.ylabel('Frequency')
    plt.title(f'train {data_list[i][1]} Histogram')


    plt.tight_layout()  
    plt.show()
```
