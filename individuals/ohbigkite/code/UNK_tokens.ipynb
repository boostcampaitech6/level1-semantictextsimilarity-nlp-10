## Token 살펴보기(원본 데이터)

```python
import pandas as pd
import transformers
import torch

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
tokenizer = transformers.AutoTokenizer.from_pretrained("snunlp/KR-ELECTRA-discriminator")
```

```python
# train data token들 살펴보기
text_columns = ['sentence_1', 'sentence_2']

outputs = []
for idx, item in train.iterrows():
    text = '[SEP]'.join([item[text_column] for text_column in text_columns])
    outputs.append(tokenizer(text, add_special_tokens=True, padding='max_length', truncation=True))
```

```python
# [UNK] token 있는 것들만 출력
for idx, item in train.iterrows():
  if '[UNK]' in outputs[idx][0].tokens:
    print(idx, outputs[idx][0].tokens)
    print(item[text_columns])
```

-> 맞춤법 검사기 API 먼저 해보는 게 좋을듯!

```python
# test data tokenizer
test_outputs = []

for idx, item in test.iterrows():
    test_text = '[SEP]'.join([item[text_column] for text_column in text_columns])
    test_outputs.append(tokenizer(test_text, add_special_tokens=True, padding='max_length', truncation=True))
```

```python
# 2번째 데이터 살펴보기 - PAD 안 보려고 75번째까지만 출력
print(test_outputs[1][0].tokens[:75])
```

```python
# [UNK] token 있는 것들만 출력
for idx, item in test.iterrows():
  if '[UNK]' in test_outputs[idx][0].tokens:
    print(idx, test_outputs[idx][0].tokens)
    print(item[text_columns])
```

```python
# 편의성을 위해 함수로 정의해버리기
def UNK(dataframe):
  text_columns = ['sentence_1', 'sentence_2']

  outputs = []

  for idx, item in dataframe.iterrows():
      text = '[SEP]'.join([item[text_column] for text_column in text_columns])
      outputs.append(tokenizer(text, add_special_tokens=True, padding='max_length', truncation=True))

  for idx, item in dataframe.iterrows():
    if '[UNK]' in outputs[idx][0].tokens:
      print(idx, outputs[idx][0].tokens[:80])
      print(item[text_columns])

  return outputs
```

```python
tr = UNK(train)
```
