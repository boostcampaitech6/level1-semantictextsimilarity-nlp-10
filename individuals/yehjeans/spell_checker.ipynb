맞춤법 검사기 호출 및 테스트

```python
from hanspell import spell_checker

sent = "맞춤법 틀리면 외 않되? 쓰고싶은대로쓰면돼지 "
spelled_sent = spell_checker.check(sent)

hanspell_sent = spelled_sent.checked
print(hanspell_sent)
```

```python
def grammar_check(val):
    
    sent = val
    spelled_sent = spell_checker.check(sent)

    hanspell_sent = spelled_sent.checked
    # print(hanspell_sent)    
        
    return hanspell_sent
```

preprocess_text -> 문장부호만 없앰. 먼저 없애줘야 맞춤법 검사기가 돌아감.

```python
import re

def preprocess_text(text):
        
        # create Korean tokenizer using soynlp library
        # tokenizer = RegexTokenizer()

        # # 2회 이상 반복된 문자를 정규화
        # text = repeat_normalize(text, num_repeats=2)
        
        # 불용어 제거
        # text = ' '.join([token for token in text.split() if not token in stopwords_2])
        # # 대문자를 소문자로 변경
        text = text.lower()
        # "<PERSON>"을 "사람"으로 변경
        #text = re.sub('<PERSON>', '사람', text)
        # 한글 문자, 영어 문자, 공백 문자를 제외한 모든 문자 제거
        text = re.sub('[^가-힣a-z\\s]', '', text)
        # # 텍스트를 토큰으로 분리  예) "안녕하세요" -> "안녕", "하", "세요"
        # tokens = tokenizer.tokenize(text)
        # 어간 추출
        # tokens = [self.stemmer.morphs(token)[0] for token in text.split()]
        # join tokens back into sentence
        
        return text
```

수정할 데이터 -> process_data 에 넣어주세요

```python

process_data = dev_data

```

맞춤법 검사기 == 띄어쓰기 
spacing 안해줘도 알아서 띄어쓰기까지 맞춤법 검사기가 해줌 개굴

```python
for idx, val in enumerate(process_data["sentence_1"]):
    # 현재 처리 중인 인덱스(idx)와 값(val)을 이용하여 전처리 수행
    processed_text = str(preprocess_text(val))
    
    
    # 전처리된 결과를 원래 데이터프레임에 대체
    process_data.at[idx, 'sentence_1'] = processed_text

# 'sentence_2' 열에 대해도 동일한 방법으로 처리 가능
for idx, val in enumerate(process_data["sentence_2"]):
    
    processed_text = str(preprocess_text(val))
    process_data.at[idx, 'sentence_2'] = processed_text
    
    
```

```python
from tqdm import tqdm
import time


for idx, val in tqdm(enumerate(process_data["sentence_1"])):
    # 현재 처리 중인 인덱스(idx)와 값(val)을 이용하여 전처리 수행
    processed_text = str(grammar_check(val))
    
    # 전처리된 결과를 원래 데이터프레임에 대체
    process_data.at[idx, 'sentence_1'] = processed_text
    time.sleep(0.01)

# 'sentence_2' 열에 대해도 동일한 방법으로 처리 가능
for idx, val in  tqdm(enumerate(process_data["sentence_2"])):
    
    processed_text = str(grammar_check(val))
    process_data.at[idx, 'sentence_2'] = processed_text
    time.sleep(0.01)
```
